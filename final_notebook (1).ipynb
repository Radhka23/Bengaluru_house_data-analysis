## GRAPH PLOTTED BETWEEN PRICE V/S LOCATION

```python
y = df['price'];
x = df['location']
plt.xlabel('location');plt.ylabel('price')
plt.scatter(x,y)
plt.rcParams['figure.figsize']= (30,20)
plt.xticks(rotation = 400)
```
